```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn import preprocessing
```

```python
train_data=np.genfromtxt("trainp2.csv",delimiter=",")
test_data=np.genfromtxt("testp2.csv",delimiter=",")
print(train_data.shape)
test_data.shape
```

```python
x_train=train_data[:,:4]
y_train=train_data[:,4]
x_test=test_data[:,:]
standard=StandardScaler().fit(x_train)
x_train=standard.transform(x_train)
x_test=standard.transform(x_test)
```

```python
def step(x_train,y_train,m,c,alp):
    m_slope=np.zeros(4)
    c_slope=0
    N=len(x_train)
    for i in range(0,7176):
        x=x_train[i]
        x=np.array(x)
        y=y_train[i]
        for j in range(0,4):
            m_slope[j]+=(-2/N)*(y-(m*x).sum()-c)*x[j]
        c_slope+=(-2/N)*(y-(m*x).sum()-c)
    for j in range(0,4):
        m[j]=m[j]-alp*m_slope[j]
    new_c=c-alp*c_slope
    return m,new_c
```

```python
def cost(x_train,y_train, m,c):
    total_cost = 0
    M = len(x_train)
    for i in range(M):
        x = x_train[i]
        x=np.array(x)
        y = y_train[i]
        total_cost += (1/M)*((y - (m*x).sum()-c )**2)
    return total_cost
```

```python
def gd(x_train,y_train,alp,noit):
    m=np.zeros(4)
    c=0
    for j in range(0,noit):
        m,c=step(x_train,y_train,m,c,alp)
        print("Cost ",j,":",cost(x_train,y_train,m,c))
    return m,c
```

```python
def run():
    alp=0.1
    noit=500
    m,c=gd(x_train,y_train,alp,noit)
    y_pred=[]
    for i in range(0,2392):
        x_i=np.array(x_test[i])
        y_pred.append((x_i*m).sum() + c)
    np.savetxt("gdproject2.csv",y_pred,delimiter=",",fmt='%0.5f')
```

```python
run()
```
